Bu kod parçası, Hugging Face'in `transformers` kütüphanesinden `AutoTokenizer` sınıfını kullanarak bir model checkpoint'i yükler ve ona karşılık gelen bir tokenizer oluşturur.

İlk satır, `transformers` kütüphanesinden `AutoTokenizer` sınıfını içe aktarır.

İkinci satır, kullanılacak modelin checkpoint adını belirtir. Bu örnekte, "distilbert-base-uncased" modeli kullanılacaktır. Bu model, küçültülmüş bir BERT modelidir ve küçük boyutu ve hızlı işlemesi nedeniyle popülerdir.

Üçüncü satır, `AutoTokenizer.from_pretrained()` fonksiyonunu kullanarak belirtilen model checkpoint'ini yükler ve bu modele karşılık gelen bir tokenizer oluşturur. `from_pretrained()` fonksiyonu, önceden eğitilmiş bir modelin veya tokenizerin adını alır ve onu yükler.

Sonuç olarak, bu kod parçası, "distilbert-base-uncased" model checkpoint'ini yükler ve bu modele karşılık gelen bir tokenizer oluşturur. Oluşturulan tokenizer, metin verilerini modele girdi olarak verebilmek için tokenlere dönüştürmek için kullanılabilir.

````python
from transformers import AutoTokenizer
model_ckpt = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
````

Bu kod parçası, verilen bir metin dizisini (text) belirtilen tokenizer kullanarak kodlar (tokenize eder) ve kodlanmış metni (encoded_text) döndürür. Sonra kodlanmış metni ekrana yazdırır.

`tokenizer(text)` çağrısı, belirtilen metin dizisini belirtilen tokenizer kullanarak kodlar. Bu işlem, metni modelin anlayabileceği sayısal tokenlere dönüştürür.

Kodlanmış metin, `encoded_text` değişkenine atanır.

`print(encoded_text)` ifadesi, kodlanmış metni ekrana yazdırır. Bu, metnin tokenizer tarafından nasıl kodlandığını görmek için kullanılır.

Ancak, dikkat etmemiz gereken bir nokta var: `tokenizer()` fonksiyonunun dönüştürdüğü sonuç bir sözlük (dictionary) türündedir. Bu sözlük, 'input_ids', 'attention_mask' ve gerekirse diğer özel durumlar gibi anahtarları içerebilir. Kodlanmış metni doğrudan ekrana yazdırmak yerine, içeriğini daha anlamlı bir şekilde görmek için sözlüğü yazdırabiliriz.


````python
text = "I love you Machine Learning! Tokenization is awesome"
encoded_text = tokenizer(text)
print(encoded_text)
````

Bu kod parçası, kodlanmış metindeki token ID'lerini geri dönüştürerek (convert_ids_to_tokens fonksiyonunu kullanarak) orijinal metindeki kelimelere (tokenlere) dönüştürür ve bu tokenleri ekrana yazdırır.

`tokenizer.convert_ids_to_tokens(encoded_text.input_ids)` çağrısı, kodlanmış metindeki token ID'lerini alır ve bu ID'leri orijinal metindeki kelimelere dönüştürür. Bu, modelin token ID'lerini gerçek kelimelere çevirerek daha anlaşılır bir çıktı elde etmemizi sağlar.

Dönüştürülen tokenler, `tokens` değişkenine atanır.

`print(tokens)` ifadesi, dönüştürülen tokenleri ekrana yazdırır. Bu şekilde, orijinal metindeki kelimeleri görebiliriz.

````python
tokens = tokenizer.convert_ids_to_tokens(encoded_text.input_ids)
print(tokens)
````

Bu kod parçası, belirli bir veri kümesinin metin örneklerini (örneğin, eğitim veri setinden ilk iki örneği) tokenize etmek için bir `tokenize` fonksiyonu tanımlar ve bu fonksiyonu çağırır.

`tokenize` fonksiyonu, `batch` adında bir parametre alır, bu parametre bir veri kümesi içindeki bir batch'i temsil eder. Fonksiyon, tokenizer'ı kullanarak metinleri tokenize eder.

İlk olarak, `tokenizer()` yöntemi çağrılır ve batch içindeki metinler (`batch["text"]`) tokenlere dönüştürülür. `padding=True` ve `truncation=True` parametreleri, tokenleştirmede dolgu (padding) ve kırpma (truncation) işlemlerinin yapılmasını sağlar.

Son olarak, tokenleştirilmiş veri, `temp` adında bir değişkene atanır ve bu değişken fonksiyon tarafından döndürülür.

Örneğin, `dataset["train"][:2]`, veri kümesinin eğitim bölümünden ilk iki örneği seçer ve bunları tokenize eder.

Kodun son kısmında, `tokenize()` fonksiyonu çağrılır ve eğitim veri setinin ilk iki örneği tokenize edilir. Sonuçlar ekrana yazdırılır.

````python
def tokenize(batch):
    temp = tokenizer(batch["text"],padding = True , truncation = True)
    return temp

print(tokenize(dataset["train"][:2]))
````

````python
print(tokenize(dataset["train"][:4]))
````

Bu kod parçası, veri setindeki tüm metin örneklerini tokenize etmek için `tokenize` fonksiyonunu kullanır ve bu işlemi veri setindeki tüm örnekler üzerinde uygular.

`dataset.map()` fonksiyonu, veri setindeki her bir öğeyi belirtilen bir fonksiyona (burada `tokenize` fonksiyonuna) uygulamak için kullanılır. Bu, veri setindeki tüm örnekleri tek tek dönüştürmek için kullanılır.

Parametreler şu şekildedir:
- `tokenize`: Her bir öğeye uygulanacak fonksiyon.
- `batched=True`: Veri seti işlemini toplu (batch) olarak yapmak için bu parametre `True` olarak ayarlanır.
- `batch_size=None`: Batch boyutu belirtilmez, böylece `map()` fonksiyonu otomatik olarak uygun bir batch boyutu belirler.

Bu kod parçası, veri setindeki her bir öğeyi tokenize etmek için `tokenize` fonksiyonunu kullanır ve tokenleştirilmiş bir veri setini döndürür. Bu işlem, her bir örnek için metinlerin tokenlere dönüştürülmesini içerir.

Sonuç olarak, `emotions_encoded` değişkeni, veri setindeki metin örneklerinin tokenlere dönüştürülmüş halini içerir. Bu tokenleştirilmiş veri seti, orijinal veri seti ile aynı yapıya sahiptir, ancak metinler tokenlere dönüştürülmüştür.

İkinci satır, `len(classes)` ifadesini kullanarak sınıf sayısını belirler. Bu, `classes` adında bir yapıda saklanan sınıf etiketlerinin (labels) sayısını döndürür. Bu sınıf etiketleri, genellikle sınıflandırma görevindeki her bir sınıfın isimlerini veya tanımlayıcılarını içerir.

Bu kod parçacığı, sınıf etiketlerinin sayısını belirledikten sonra, bir çok sınıflı sınıflandırma (multi-class classification) modeli oluşturmak için `AutoModelForSequenceClassification` sınıfını kullanmak üzere bu sayıyı kullanabilirsiniz. Bu kod parçası, mevcut cihazın GPU destekli olup olmadığını kontrol eder ve ardından belirtilen model_checkpoint'ten önceden eğitilmiş bir sınıflandırma modeli (`AutoModelForSequenceClassification`) yükler. Modeli GPU veya CPU'ya taşıyarak uygun cihazda çalışmasını sağlar.

Sonuç olarak, `model` değişkeni, belirtilen model_checkpoint'ten yüklenen ve belirtilen cihaza taşınan bir sınıflandırma modelini temsil eder. Bu model, metin dizilerinin sınıfını tahmin etmek için kullanılabilir.

Bu kod parçacığı, Hugging Face Transformers kütüphanesinden `TrainingArguments` sınıfını içe aktarır ve ardından eğitim işlemi için gerekli olan argümanları belirler.

`TrainingArguments` sınıfı, eğitim işlemi sırasında kullanılacak argümanları tutar ve yapılandırır. Bu argümanlar, eğitim sürecini yönetmek için kullanılır, örneğin, eğitim süresi, batch boyutları, öğrenme oranı, değerlendirme stratejisi vb.

`TrainingArguments` sınıfının parametreleri şu şekildedir:
- `output_dir`: Eğitim sürecinde oluşturulan model dosyalarının kaydedileceği dizin.
- `num_train_epochs`: Eğitim sürecinde kullanılacak toplam epoch sayısı.
- `learning_rate`: Öğrenme oranı.
- `per_device_train_batch_size`: Her bir eğitim cihazı için kullanılacak eğitim batch boyutu.
- `per_device_eval_batch_size`: Her bir değerlendirme cihazı için kullanılacak değerlendirme batch boyutu.
- `weight_decay`: Ağırlık azalma (weight decay) katsayısı.
- `evaluation_strategy`: Değerlendirme stratejisi ("epoch" veya "steps").
- `disable_tqdm`: TQDM ilerleme çubuğunun devre dışı bırakılıp bırakılmayacağını belirten bir boolean değer.

Bu parametreler, eğitim sürecinin nasıl yapılandırılacağını belirler ve eğitim sırasında kullanılır.

Örnekte, belirli bir modelin eğitimi için kullanılacak argümanlar `training_args` adında bir değişkene atanır. Bu argümanlar, belirtilen eğitim ayarlarını temsil eder ve eğitim işlemi sırasında kullanılır.

````python
from transformers import TrainingArguments
batch_size = 64
model_name = "distilbert-finetuned-emotion"
training_args = TrainingArguments(output_dir = model_name,
                                  num_train_epochs = 2,
                                  learning_rate = 2e-5,
                                  per_device_train_batch_size = batch_size,
                                  per_device_eval_batch_size = batch_size,
                                  weight_decay = 0.01,
                                  evaluation_strategy = "epoch",
                                  disable_tqdm = False)
````

Bu kod parçası, sklearn kütüphanesinden `accuracy_score` ve `f1_score` fonksiyonlarını içe aktarır ve ardından tahmin edilen etiketler ve gerçek etiketler arasındaki doğruluk (accuracy) ve F1 puanını hesaplamak için bir `compute_metrics` fonksiyonunu tanımlar.

`compute_metrics` fonksiyonu, `pred` parametresi olarak bir tahmin nesnesi alır. Bu tahmin nesnesi, modelin bir veri kümesi üzerindeki tahminlerini ve gerçek etiketleri içerir.

Fonksiyonun içindeki işlemler şu adımları takip eder:

1. `pred.label_ids` ifadesi, tahmin nesnesindeki gerçek etiketleri alır.
2. `pred.predictions.argmax(-1)` ifadesi, tahmin nesnesindeki tahmin edilen etiketleri alır. `argmax(-1)` işlemi, her bir örnek için en yüksek olasılığa sahip etiketi seçer.
3. `f1_score(labels, preds, average="weighted")` ifadesi, gerçek ve tahmin edilen etiketler arasındaki F1 puanını hesaplar. `"weighted"` ortalama kullanılarak her bir sınıfın ağırlıklı F1 puanı hesaplanır.
4. `accuracy_score(labels, preds)` ifadesi, gerçek ve tahmin edilen etiketler arasındaki doğruluğu hesaplar.

Fonksiyon, hesaplanan doğruluk ve F1 puanlarını bir sözlük olarak döndürür. Bu sözlük, `accuracy` ve `f1` anahtarlarına sahip ve bu değerleri içerir. Bu değerler, modelin performansını ölçmek için kullanılabilir. Bu özellikle, sınıflandırma görevlerinde modelin başarısını değerlendirmek için yaygın olarak kullanılan metriklerdir.

````python
from sklearn.metrics import accuracy_score,f1_score

def compute_metrics(pred):
  labels = pred.label_ids
  preds = pred.predictions.argmax(-1)
  f1 = f1_score(labels,preds,average = "weighted")
  acc = accuracy_score(labels,preds)
  return {"accuracy":acc,"f1":f1}
````

Bu kod parçası, Hugging Face Transformers kütüphanesinden `Trainer` sınıfını içe aktarır ve ardından bir `Trainer` nesnesi oluşturur.

`Trainer` sınıfı, bir modelin eğitimini ve değerlendirmesini yönetmek için kullanılır. Bu sınıf, eğitim parametrelerini ve metrik hesaplama fonksiyonunu içeren bir dizi argümanı alır.

`Trainer` nesnesinin oluşturulması için kullanılan parametreler şunlardır:

- `model`: Eğitilecek ve değerlendirilecek model.
- `args`: Eğitim argümanları (`TrainingArguments` nesnesi).
- `compute_metrics`: Metrik hesaplama fonksiyonu (`compute_metrics` fonksiyonu).
- `train_dataset`: Eğitim veri seti.
- `eval_dataset`: Değerlendirme veri seti.
- `tokenizer`: Modelin girdi verilerini işlemek için kullanılan tokenizer.

Bu parametreler, eğitim ve değerlendirme işlemlerini yönetmek için gerekli olan tüm bileşenleri sağlar.

Örneğin, `trainer` değişkeni, belirtilen modelin eğitimini ve değerlendirmesini yönetmek için kullanılacak bir `Trainer` nesnesini temsil eder. Bu nesne, belirtilen eğitim parametreleri ve veri setleri ile birlikte kullanılabilir, böylece modelin eğitimi ve performansının değerlendirilmesi sağlanabilir.

````python
from transformers import Trainer

trainer = Trainer(model = model,
                  args = training_args,
                  compute_metrics = compute_metrics,
                  train_dataset = emotions_encoded["train"],
                  eval_dataset = emotions_encoded["validation"],
                  tokenizer = tokenizer )
````

Eğitim süreci boyunca, modelin her bir epoch'taki kaybı, doğruluğu ve diğer metrikler gibi istatistikleri `trainer` nesnesi tarafından takip edilir ve görselleştirilir. Ayrıca, eğitim sırasında her bir epoch'ta modelin performansı da değerlendirilir ve bu bilgiler ekrana yazdırılır.
